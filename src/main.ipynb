```python
elapsed = 7835
hours = elapsed // 3600
minutes = (elapsed - (hours * 3600)) // 60
seconds = elapsed - ((hours * 3600) + (minutes * 60))
print(hours, minutes, seconds)
```

```python
elapsed = 7835
hours = elapsed // (60 * 60)
minutes = (elapsed % (60 * 60)) // 60
seconds = (elapsed % (60 * 60)) % 60

print(hours, minutes, seconds)
```

```python
score = 700
rating = "Invalid"
if score < 0:
	rating = "Negative"
elif score < 580:
	rating = "Poor"
elif score < 670:
	rating = "Fair"
elif score < 740:
	rating = "Good"
elif score < 800:
	rating = "Very Good"
elif score < 850:
	rating = "Excellent"
print(rating)
```

```python
elapsed = 1000000
magnitude = "Invalid"
if elapsed < 60:
	magnitude = "seconds"
elif elapsed < (60 * 60):
	magnitude = "minutes"
elif elapsed < (60 * 60 * 24):
	magnitude = "hours"
elif elapsed < (60 * 60 * 24 * 7):
	magnitude = "days"
elif elapsed > (60 * 60 * 24 * 7):
	magnitude = "weeks"
print(magnitude)
```
